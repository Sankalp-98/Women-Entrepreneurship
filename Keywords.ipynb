```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer as SIA
```

```python
score=[]
for i in range(len(df)):
    score.append(SIA().polarity_scores(df['Brief about your start-up '][i]))
score
```

```python
df['score']=pd.DataFrame(score)['compound']
df
```

```python
Sentiment=[]
for i in range(len(df)):
    if df['score'][i]==0:
        Sentiment.append('Neutral')
    elif df['score'][i]>0:
        Sentiment.append('Positive')
    else:
        Sentiment.append('Negative')
Sentiment
```

```python
df['Sentiment']=Sentiment
df.head()
```
